```python
# copy all .pngs in subfolders of location D:\Research2021\Rad_TM\code\Baselinevscontrols to a single folder

import os
import shutil
import glob

# location of the subfolders
#location = r'D:\Research2021\Rad_TM\data\pat data-20231016T160300Z-001\controls\T1'
location = r'D:\Research2021\Rad_TM\data\pat data-20231016T160300Z-001\pat data\done'

# location of the folder to move the .pngs to
destination = r'D:\Research2021\Rad_TM\code\Baselinevscontrols\all_pngs'

# loop through all subfolders
for file in glob.glob(location + '/*/Baseline/*.png'):
            print(file)
            # rename the file to include the subfolder name
            file_name = file.split("\\")[-3].split("_")[0]
            print(file_name)
            file_n =  "baseline_" + file_name + "_" + file.split("_")[-1]
            print(file_n)
            try:
                os.rename(file, os.sep.join(file.split(os.sep)[0:-1], file_n))
            except:
                pass
            shutil.copyfile(file, os.path.join(destination,file_n))
            print(f'{file} moved to {destination}')
```
